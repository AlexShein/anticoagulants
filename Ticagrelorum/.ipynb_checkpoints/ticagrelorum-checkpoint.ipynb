# Initial analysis

```python
UNNEEDED_COLUMNS = []
```

```python
df = pd.read_csv('ticagrelorum.csv', header=0, index_col=0).reset_index(drop=True)
```

```python
def is_empty(col):
    return all(df[col].isna()) or all(df[col] == ' ')
EMPTY_COLUMNS = [col for col in df.columns if is_empty(col)]
EMPTY_COLUMNS
```

```python
SINGLE_VALUE_COLUMNS = [col for col in df.columns if len(df[col].unique()) == 1]
SINGLE_VALUE_COLUMNS
```

```python
df = df.drop([*UNNEEDED_COLUMNS, *SINGLE_VALUE_COLUMNS, *EMPTY_COLUMNS], axis=1)
```
